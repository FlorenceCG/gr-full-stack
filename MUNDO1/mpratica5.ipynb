```python
from faker import Faker
from random import randint  
import matplotlib.pyplot as plt
from num2words import num2words
from wordcloud import WordCloud

fake = Faker('pt_BR') 

```

```python
arquivo = open('AlunosPontos.txt','w+')
listaNum = []
for i in range(100):
  Nome = fake.name()
  Notas = str(fake.random_int(1,100))
  arquivo.writelines(Nome)
  arquivo.writelines(',')
  arquivo.writelines('\n')
  print(Nome, Notas)
  listaNum.append(Notas)

arquivo.close()

  
```

```python
arquivo = open('AlunosPontos.txt','r')
arquivoS = []
for linha in arquivo:
  linha_limpa = linha.strip()
  arquivoS.append(linha_limpa)

arquivo.close()

```

```python
strnum = []
intnum = []

for linha in listaNum:
  intnum.append(int(linha))
  strnum.append(num2words(int(linha),lang='pt,br'))

```

```python
n, bins, colunas = plt.hist(intnum, facecolor='green', alpha=0.80);
plt.xlabel('Valores')
plt.ylabel('Probabilidade')
plt.title('Histograma dos valores')
plt.grid(True)
plt.show
```

```python
nuvem = WordCloud(background_color='grey',
                           width=800, height=400).generate(str(strnum))
plt.imshow(nuvem, interpolation='bilinear')
plt.axis('off')
nuvem.to_file('Nuvem de palavras.png')
```
